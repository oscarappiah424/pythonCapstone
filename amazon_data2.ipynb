From the data, it can be seen that a currency sign is attached to the values in the ***discounted_price and actual_price column***.

```python
# Replacing currency sign and comma sign 

df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')


#Changing the data type of discounted price and actual price

df['discounted_price'] = df['discounted_price'].astype('float64')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
#Changing Datatype and values in Discount Percentage

df['discount_percentage'] = df['discount_percentage'].str.replace('%','')
df['discount_percentage'] = df['discount_percentage'].astype('float64')
df['discount_percentage'] = df['discount_percentage'] / 100

df['discount_percentage']
```

```python
#Replacing the unwanted string with 4.0 and changing Rating Columns Data Type

df['rating'] = df['rating'].str.replace('|', '4.0')
```

From the category column in the table above, it can be seen that the names are not well sorted.

```python
#Creating a new DataFrame with Selected Column

df1 = df[['product_id', 'product_name', 'category', 'discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].copy()
```

```python
#Splitting the Strings in the category column

catsplit = df['category'].str.split('|', expand=True)
catsplit
```

```python
#Renaming category column

catsplit = catsplit.rename(columns={0:'category_1', 1:'category_2', 2:'category_3'})
catsplit
```

```python
#Adding categories to the new dataframe

df1['category_1'] = catsplit['category_1']
df1['category_2'] = catsplit['category_2']
df1['category_3'] = catsplit['category_3']

df1
```

#### Since the category names has been sorted, the main category can be dropped

```python
# Dropping the the main category column

df1.drop(columns='category', inplace=True)

df1
```

```python
#Checking category_1 unique values

df1['category_1'].value_counts()
```

```python
#Fixing Strings in the Category_1 Column

df1['category_1'] = df1['category_1'].str.replace('&', ' & ')
df1['category_1'] = df1['category_1'].str.replace('OfficeProducts', 'Office Products')
df1['category_1'] = df1['category_1'].str.replace('MusicalInstruments', 'Musical Instruments')
df1['category_1'] = df1['category_1'].str.replace('HomeImprovement', 'Home Improvement')
```

```python
#Checking category_1 unique values again

df1['category_1'].value_counts()
```

```python
#Checking category_2 unique values

df1['category_2'].value_counts()
```

```python
#Fixing Strings in Category_3 column

df1['category_3'] = df1['category_3'].str.replace('&', ' & ')
df1['category_3'] = df1['category_3'].str.replace(',', ', ')
df1['category_3'] = df1['category_3'].str.replace('ExternalMemoryCardReaders', 'External MemoryCard Readers')
df1['category_3'] = df1['category_3'].str.replace('SmallKitchenAppliances', 'Small Kitchen Appliances')
df1['category_3'] = df1['category_3'].str.replace('MobileAccessories', 'Mobile Accessories')

df1['category_3'].value_counts()
```

From the statistical data above, the mean rating is approximately 4

```python
#Creating Rankings for the ratings as rating_cat

rating_cat = []

for i in df1['rating']:
    if i < 2.0 : rating_cat.append('Poor')
    elif i < 3.0 : rating_cat.append('Below Average')
    elif i < 4.0 : rating_cat.append('Average')
    elif i < 5.0 : rating_cat.append('Above Average')
    elif i == 5.0 : rating_cat.append('Excellent')
```

```python
# Adding the rating_cat Column and Changing the Data Type

df1['rating_cat'] = rating_cat

df1['cat'] = df1['rating_cat'].astype('category')
```
